```python
df=df.rename(columns={"student_id":"ID","name":"Name","age":"Age","gender":"Gender","grade":"Class","math_score":"Math_score","english_score":"English_Score"})
df
```

```python
df["Gender"]=df["Gender"].replace({"male":"M","female":"F","Male":"M","FEMALE":"F","MALE":"M","Female":"F"})
df
```

```python
df["Class"]=df["Class"].replace({"10th":"X","11th":"XI","12th":"XII","10":"X","11":"XI","12":"XII"})
df
```

```python
df["Age"]=df["Age"].fillna(0).astype(int)
df
```

```python
df["Math_score"]=df["Math_score"].fillna(0).astype(int)
df
```

```python
df["English_SCore"]=df["English_Score"].replace({"missing":"0"})
df
```

```python
df["English_Score"]=df["English_Score"].fillna(0).astype(int)
print(df.head())
```

```python
df["Name"]=df["Name"].str.strip()
df
```

```python
df[["First_Name","Last_Name"]]=df["Name"].str.split(" ",n=1,expand=True)
df.drop(columns=["Name"],inplace=True)
df
```

```python
df["remarks"]=df["remarks"].replace({"excellent":"A","good":"B","GOOD":"B","Average":"C","average":"C","poor":"D","needs improvement":"C","good student":"B"})
df
```

```python
df["enrolled_date"]=df["enrolled_date"].astype(str).str.replace("/","_",regex=True)
df
```

```python
df["Last_Name"]=df["Last_Name"].str.strip(",")
df
```

```python
df=df[
    [
        "ID",
        "First_Name",
        "Last_Name",
        "Age",
        "Gender",
        "Class",
        "Math_score",
        "English_Score",
        "science_score",
        "enrolled_date",
        "remarks"

    ]
]
df
```

```python
df['enrolled_date']=pd.to_datetime(
    df['enrolled_date'],
    dayfirst=True,
    errors='coerce'
).dt.strftime('%d-%m-%y').fillna(df['enrolled_date'])
df
```
